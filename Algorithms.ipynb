```python
#Bubble sort
li=[89,75,98,7,56]
n=len(li)
for i in range(0,n-1):
  for j in range(0,n-i-1):
    if(li[j]<li[j+1]):
      t=li[j]
      li[j]=li[j+1]
      li[j+1]=t
print(li)
```

```python
#finding maximum element in a finite sequence
li=[89,75,98,7,56]
n=len(li)
x=li[0];
for i in range(1,n):
  if(x<li[i]):
    x=li[i]
print(x)

```

Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
 Output: Because nums[0] + nums[1] == 9, we return [0, 1].

```python
list_nums=[3,2,3,17,9,6]
target=23
n=len(list_nums)
for i in range(0,n):
  for j in range(i+1,n):
    if list_nums[i] + list_nums[j] == target:
      print('the indices=',[i,j], ' , target=',target)
#the above code has time = O(n^2), space = O(1).     Therefore we can improve our time to O(n) . 
#improve code below

Hashmap ={}
for i in range(n):
  var = target - list_nums[i]
  if var in Hashmap: 
    print([i, Hashmap[var]])
  Hashmap[list_nums[i]] = i  
```

Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

Assume the environment does not allow you to store 64-bit integers (signed or unsigned).

```python
class Solution:
    def reverse(self, x: int) -> int:
        nn=abs(x)
        nnum = 0
        while nn>0:
            last_digit = nn%10
            nnum = nnum*10 + last_digit 
            nn = nn//10
        if x<0:
            nnum =-nnum
        else:
            nnum = abs(nnum)
        if nnum < -2**31 or nnum > (2**31)-1:
          nnum = 0
        else:
          nnum = nnum
        return nnum
temp = Solution()    
print(temp.reverse(1986))   

#recursive as class
class Solution:
    def reverse(self, x: int) -> int:
      if x < 10:
         print(x) 
      else:
        print(x%10,end = '')
        Solution().reverse(x//10)
temp1 = Solution()    
temp1.reverse(15233)          
       
```
